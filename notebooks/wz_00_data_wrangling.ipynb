# Creating work zones and getting their total length
### DOT Street Closure data identifies locations in the New York City Street Closure map where a street is subject to a full closure, restricting through traffic, for the purpose of conducting construction related activity on a City street. Details about DOT construction permits can be found at Street Works Manual, http://streetworksmanual.nyc/. Full Closure Permits are issued for a period of time during which the street may be closed to through traffic for only a portion of the time, and open at other times.
### Multiple records in the street closures data show the same segment id and street location, however different purposes, dates, and times. The difference in date and time is sometimes only seconds or minutes. These records describe the same work zone event by listing the different activities happening in it through a minor time change. We want to identify these records and group them so that we get one record for each street closure event at the same location and date/time. After that we aim at joining contiguous segments that witness work at the same time. These would describe work zones that include multiple segments closures.

### 2.5 Group continuous dates as one work zone
### Testing with the sample

```python
# taking a sample
test = expand_2.loc[expand_2['segmentid']==9017006]
test.head()
```

```python
# concatenating the list of purposes in each row
test['purpose'] = test['purpose'].apply(lambda x:np.concatenate(x).ravel())

# counting the days between each date entry 
test['diff'] = test['date'].diff()
test['diff'].iloc[0] = 0
test['diff'].iloc[1:] = test['diff'].iloc[1:].apply(lambda x:x.days)

test.reset_index(inplace=True) 
test.head()
```

```python
# consider consecutive days of work on the same segment as one work zone event
# locate records where the difference between consecutive dates is more than one
# a difference more than one would mean that work was interrupted on the segment
# and therefore we will consider this record the start of a new work zone event 
test_split_point = test.loc[test['diff']>1].index.tolist()
test_split_point
```

```python
previous_ = 0
list_split_dataframe = []
for i in test_split_point:
    list_split_dataframe.append(test.loc[previous_:i-1])
    previous_ = i
```

```python
list_split_dataframe[0]
```

```python
list_split_dataframe[1]
```

```python
# creating a template for the new work zones data frame we will generate
wz_template = pd.DataFrame(index=[0], columns=['segmentid','created_date','closed_date','purpose'])
wz_template
```

```python
# grouping consecutive dates using the split points created above
list_df_wz = []
for df in list_split_dataframe:
    wz_template_temp = wz_template.copy()
    wz_template_temp['segmentid'] = 9017006
    wz_template_temp['created_date'] = df['date'].min()
    wz_template_temp['closed_date'] = df['date'].max()
    wz_template_temp.at[0,'purpose'] = np.unique(np.concatenate(df['purpose'].to_numpy()).ravel()).tolist()
    list_df_wz.append(wz_template_temp)

pd.concat(list_df_wz, axis=0, ignore_index=True)
```

### Applying on the whole data

```python
expand_3 = expand_2.copy()

# concatenating the list of purposes in each row
expand_3['purpose'] = expand_3['purpose'].apply(lambda x:np.concatenate(x).ravel())

# list of all the segment ids to loop over 
list_segment_id = expand_3['segmentid'].unique()
```

```python
i=0
list_wz_processed = []
for s in list_segment_id:
    df_target = expand_3.loc[expand_3['segmentid']==s]
    df_target['diff'] = df_target['date'].diff()
    df_target['diff'].iloc[0] = 0
    df_target['diff'].iloc[1:] = df_target['diff'].iloc[1:].apply(lambda x:x.days)
    df_target = df_target.reset_index() 
    list_split_point = df_target.loc[df_target['diff']>1].index.tolist()
    previous_ = 0
    list_split_dataframe = []
    if len(list_split_point)>0:
        for j in list_split_point:
            list_split_dataframe.append(df_target.loc[previous_:j-1])
            previous_ = j
    else:
        list_split_dataframe.append(df_target)
        
    list_df_wz = []
    for df in list_split_dataframe:
        wz_template_temp = wz_template.copy()
        wz_template_temp['segmentid'] = s
        wz_template_temp['created_date'] = df['date'].min()
        wz_template_temp['closed_date'] = df['date'].max()
        wz_template_temp.at[0,'purpose'] = np.unique(np.concatenate(df['purpose'].to_numpy()).ravel()).tolist()
        list_df_wz.append(wz_template_temp)
    list_wz_processed.append(pd.concat(list_df_wz, axis=0, ignore_index=True)) 
```

### 3.3 Applying to the whole data

```python
def create_workzones(df):  
    ''' a function that takes a group of segment closures within the same time frame, group contiguous segments, 
    and assign them to a distinct work zone'''
    sep_wz = gpd.GeoDataFrame(geometry=gpd.GeoSeries(unary_union(df.geometry.buffer(0.1))))
    sep_wz.crs = df.crs
    wz_id = [str(df.time_frame.unique()[0]) +  '_' + str(i) for i in range(len(sep_wz))]
    sep_wz['wz_id'] = wz_id
    df = gpd.sjoin(df, sep_wz, how='left').drop('index_right', 1)
    return df
```

```python
# testing function
create_workzones(test_2)
```

```python
# applying the function to all the records (wz_geo)
wz_sep = wz_geo.groupby('time_frame').apply(lambda x: create_workzones(x))
print(wz_sep.shape)
wz_sep.head()
```

### 4. Adding duration and season attributes

```python
# calculing total durations of generated work zones 
wz_sep['duration_days'] = (wz_sep.closed_date - wz_sep.created_date).dt.days
wz_sep.head(2)
```

```python
def getSeason(dates):
    ''' takes a series of dates and returns a series of seasons'''
    seasons = []
    for i, date in dates.items():
        month = date.month
        if (month > 11 or month <= 3):
           seasons.append('winter')
        elif (month == 4 or month == 5):
           seasons.append('spring')
        elif (month >=6 and month <= 9):
           seasons.append('summer')
        else:
           seasons.append('fall')
    return pd.Series(seasons)
```

```python
# getting the season of the start and end date of generated work zones
wz_sep['created_season'] = getSeason(wz_sep.created_date)
wz_sep['closed_season'] = getSeason(wz_sep.closed_date)

wz_sep.head(2)
```

### 5. Grouping records with the same generated work zone id

```python
# grouping by work zone id, and aggregating
# line geometries representing each segment will be agggregated into a Multiline string

work_zones = wz_sep.groupby('wz_id').agg({'segmentid': 'unique',
                                        'created_date': 'first',
                                        'closed_date': 'first',
                                        'created_season': 'first',
                                        'closed_season': 'first',
                                        'duration_days': 'first',
                                        'purpose': 'unique',
                                        'geometry': lambda x: geometry.MultiLineString(x.tolist())}).reset_index()
print(work_zones.shape)
work_zones.head()
```

```python
work_zones.info()
```

```python
# calculating the length in feet of the closed segments within each generated work zone 
work_zones['length_ft'] = pd.Series([l.length for i, l in work_zones.geometry.items()])
work_zones.head()
```
