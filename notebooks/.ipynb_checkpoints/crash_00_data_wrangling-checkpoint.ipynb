## Motor Vehicle Collisions Cleaning and EDA
### This notebook cleans collisions data set. Collisions data will be used to calculate the number of crashes in proximity to a work zone. Also, after clustering the work zones the number of collisions will be used to calculate the probability of a crash in a cluster. 

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import shapely
from shapely.geometry import Point, Polygon
import geopandas as gpd
from fiona.crs import from_epsg

import warnings
warnings.filterwarnings('ignore')

%pylab inline
pylab.rcParams['figure.figsize'] = 12, 12
pylab.rcParams['figure.dpi'] = 100

pd.set_option("display.max_columns", 100)
pd.set_option("display.min_rows", 100)
```

### Importing Data

```python
# data can be downloaded from: https://data.cityofnewyork.us/Public-Safety/Motor-Vehicle-Collisions-Crashes/h9gi-nx95
# data used was downloaded in March 2020
data = pd.read_csv('../data/collisions/Motor_Vehicle_Collisions_-_Crashes.csv', parse_dates=['CRASH DATE'])
```

```python
print("Data Shape: {}\n\nNull Values in each column\n".format(data.shape))
for col in data.columns:
    print(col, '\t', data[col].isnull().sum())
```

```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(" ","_")
```

```python
data.head()
```

```python
data.info()
```

```python
for col in data.columns:
    print()
    print(col, data[col].unique())
```

```python
data['zip_code'] = pd.to_numeric(data['zip_code'], errors='coerce')
```

```python
data['zip_code'] = data['zip_code'].fillna(-123)
data['zip_code'] = data['zip_code'].astype('int')
data['zip_code'] = data['zip_code'].replace(-123, np.nan)
```

```python
# replacing blank cells with nans
data = data.replace(r'^\s+$', np.nan, regex=True)
```

```python
data.info()
```

```python
data.head()
```

```python
strcols = ['borough', 'on_street_name', 'cross_street_name', 'off_street_name','contributing_factor_vehicle_1', 'contributing_factor_vehicle_2', 'contributing_factor_vehicle_3', 'contributing_factor_vehicle_4', 'contributing_factor_vehicle_5']
for strcol in strcols:
    data[strcol] = data[strcol].str.lower()
```

```python
crash.head()
```

```python
# dropping  null values
crash.dropna(subset=["latitude", "longitude"], inplace=True)
print(crash.shape)
# removing crashes that are mapped to 0,0
crash = crash[(crash.longitude != 0) & (crash.latitude != 0)]
print(crash.shape)
```

```python
# combine lat and lon to one column
crash['lonlat'] = list(zip(crash.longitude, crash.latitude))
# Create Point Geometry for based on lonlat column
crash['geometry'] = crash[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
print(type(crash))
crash.head(3)
```

```python
crash = crash.drop('lonlat', axis=1)
```

```python
crash = gpd.GeoDataFrame(crash)
print(type(crash))
```

## filter according to 511events dates

```python
# filter by datetime values greater or equal 2016-10-01
# 511 events data starts from 2016-10-01
crash_511 = crash_nyc.loc[crash_nyc['crash_date'] >= np.datetime64('2016-10-01')].sort_values(by='crash_date').reset_index(drop=True)

print(type(crash_511))
crash_511.head(3)
```

## filter according to street closures dates

```python
# filter by datetime values greater or equal 2017-02-02
# work zones data starts from 2017-02-02
crash_wz = crash_nyc.loc[crash_nyc['crash_date'] >= np.datetime64('2017-02-02')].sort_values(by='crash_date').reset_index(drop=True)

print(type(crash_wz))
crash_wz.head(3)
```
