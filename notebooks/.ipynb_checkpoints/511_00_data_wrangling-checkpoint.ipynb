## 511 Events - Data Cleaning & Filtering
### This notebook cleans 511 event data to only construction events for responding organizations that operate on surface roads and bridges. These 511 events will go on to be matched to a SharedStreets ID and the basis for work zone calculations

### Importing Packages and Data

```python
import numpy
import pandas as pd
import geopandas as gpd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns

pylab.rcParams['figure.figsize'] = 30,30
pylab.rcParams['figure.dpi'] = 100
```

```python
# please download raw data from: https://data.ny.gov/Transportation/511-NY-Events-Beginning-2010/ah74-pg4w
# data used was downloaded in April 2020
events = pd.read_csv('../data/street_closures/511_NY_Events__Beginning_2010.csv')
```

```python
# Converting columns to datetime values
events['Create Time'] = pd.to_datetime(events['Create Time']).copy()
events['Close Time'] = pd.to_datetime(events['Close Time']).copy()
```

```python
events.head(1)
```

### Filtering and Cleaning

```python
#FILTERING OUT RECORDS FROM BEFORE 2016
events = events[events['Close Time'] > '12/31/2016']
```

```python
#FILTERING FOR NEW YORK STATE ONLY
events = events[events['State']== 'NY']
```

```python
#FILTERING FOR ONLY  KINGS, BRONX, QUEENS, NEW YORK COUNTIES
counties = ['New York', 'Queens', 'Bronx', 'Kings']
events = events[events['County'].isin(counties)]
events.shape
```

```python
events['Responding Organization Id'].value_counts()
```

```python
#FILTERING ONLY FOR RESPONDING ORGANIZATIONS THAT MAKE SENSE - WE ONLY WANT CONSTRUCTION EVENTS ON SURFACE ROADS AND BRIDGES
organizations = ['New York State Department of Transportation (Region 11)',
                 'Port Authority New York/New Jersey',
                 'MTA Bridges and Tunnels',
                 'New York City Police Department Traffic Management Center (Long Island City)',
                 'New York State Department of Transportation (Region 10)',
                 'New York State Thruway-State Wide Operations Center - Albany',
                 'TRANSCOM',
                 'New York City Department of Transportation']

events = events[events['Responding Organization Id'].isin(organizations)]
```

```python
#CALCULATING THE DURATION OF A 511 EVENT FORM CREATE TIME TO CLOSE TIME
events['Duration'] = events['Close Time'] - events['Create Time']
```

### Exporting to CSV and Shapefile

```python
#EXPORTING TO CSV
events.to_csv('../data/cleaned_data/511_events_cleaned.csv')
```

```python
#CREATING SHAPEFILE OF POINTS
#Shapefiles cannot handle datetime format - converting columsn to strings
events['Create Time'] = events['Create Time'].astype('str')
events['Close Time'] = events['Close Time'].astype('str')
events['Duration'] = events['Duration'].astype('str')


events_gpd = gpd.GeoDataFrame(
    events, geometry=gpd.points_from_xy(events.Longitude, events.Latitude))
```

```python
#THERE WERE SOME POINTS THAT HAD INCORRECT LONGITUTDES. THIS GETS RID OF THEM.
events_gpd = events_gpd[events_gpd.Longitude < 0]
```

```python
events_gpd.plot()
```

```python
#EXPORTING TO SHAPEFILE
events_gpd.to_file('../data/cleaned_data/511_events_cleaned.shp')
```
